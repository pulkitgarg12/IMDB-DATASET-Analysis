```python
#Classify Movies Based on Rating [Excellent,Good and Average]

def rating(rating):
  if rating>=7.0:
    return "Excellent"
  elif rating>=7.0:
    return "Good"
  else:
    return "Average"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

```python
# Find Unique Values From Genre
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
  list1.append(value.split(','))
list1
```

```python
#convert 2d list int 1d
one_d=[]
for item in list1:
  for item1 in item:
    one_d.append(item1)
one_d
```

```python
uni_list=[]
for item in one_d:
  if item not in uni_list:
      uni_list.append(item)
uni_list
```
